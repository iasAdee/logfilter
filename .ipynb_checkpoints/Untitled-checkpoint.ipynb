````python
import cv2
import numpy as np
import easyocr
import cv2

LANG = ["en","de"]  # German
reader = easyocr.Reader(LANG)
def read_text(file_path=None, image_bytes=None):
    
    """
    OCR image from file path or bytes
    Returns text, per-word confidence, and overall confidence
    """

    if file_path:
        img = cv2.imread(file_path)

    elif image_bytes:
        nparr = np.frombuffer(image_bytes, np.uint8)
        img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

    else:
        return {"error": "No image source provided"}

    if img is None:
        return {"error": "Unable to read image"}

    result = reader.readtext(img)

    words = []
    full_text = []
    weighted_conf = 0
    total_len = 0

    for bbox, text, conf in result:
        text_clean = text.strip()
        if not text_clean:
            continue

        length = len(text_clean)

        words.append({
            "text": text_clean,
            "confidence_%": round(conf * 100, 2),
            #"bbox": bbox
        })

        full_text.append(text_clean)
        weighted_conf += conf * length
        total_len += length

    overall_confidence = (
        round((weighted_conf / total_len) * 100, 2)
        if total_len > 0 else 0.0
    )

    return {
        "Plain_text": " ".join(full_text),
        "Overall_confidence_%": overall_confidence,
        "Words": words
    }
````

````python
import io
import pytesseract

def extract_text_tesseract(image_input):
    """
    Extract text from image - accepts either file path or bytes
    """
    try:
        # If input is bytes, convert to PIL Image
        if isinstance(image_input, bytes):
            image = Image.open(io.BytesIO(image_input))
        # If input is string (file path)
        elif isinstance(image_input, str):
            image = Image.open(image_input)
        # If already a PIL Image
        elif isinstance(image_input, Image.Image):
            image = image_input
        else:
            raise ValueError("Input must be file path, bytes, or PIL Image")
        
        # Configure for German and English
        custom_config = r'--oem 3 --psm 6 -l deu+eng'
        text = pytesseract.image_to_string(image, config=custom_config)
        
        return text
        
    except Exception as e:
        print(f"Error extracting text: {e}")
        return ""
````

````python
def display_crops(cropped_regions):
    n = len(cropped_regions)
    if n == 0:
        print("No crops found!")
        return
    
    # Create subplots
    cols = min(3, n)  # Max 3 columns
    rows = (n + cols - 1) // cols
    
    fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
    
    # Handle single crop case
    if n == 1:
        axes = [axes]
    else:
        axes = axes.flatten()
    
    for i, crop in enumerate(cropped_regions):
        # Convert BGR to RGB for display
        crop_rgb = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)
        axes[i].imshow(crop_rgb)
        axes[i].set_title(f"Crop {i+1}")
        axes[i].axis('off')
    
    # Hide empty subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()
````

````python
def get_lines(text):
    return [l.strip() for l in text.splitlines() if len(l.strip()) > 3]

PRODUCT_NAME_REGEX = re.compile(
    r'''
    ^                                   # start of line
    (?=.*[a-zA-Z])                      # must contain letters
    (?=.*\d)                            # must contain numbers
    [a-zA-Z0-9®\-/ ]{6,60}              # allowed chars
    $                                   # end of line
    ''',
    re.VERBOSE
)

def score_name(line):
    score = 0

    if len(line.split()) <= 5:
        score += 2

    if any(char.isdigit() for char in line):
        score += 2

    if '-' in line:
        score += 1

    if line.isupper():
        score += 1

    if 'un' not in line.lower():
        score += 1

    return score



def clean_product_name(name: str) -> str:
    name = name.strip()

    # Remove trailing lowercase garbage tokens
    name = re.sub(
        r'(\d[\w\-]*)\s+[a-z]{2,4}$',
        r'\1',
        name
    )

    # Normalize spaces
    name = re.sub(r'\s+', ' ', name)
    
    return name

def clean_product_prefix(name: str) -> str:
    name = name.strip()

    # Remove short leading junk tokens (Dx, Rx, >, *, etc.)
    name = re.sub(
        r'^(?:[^\w]{0,2})?[a-zA-Z]{1,3}\s+(?=[A-Z][a-zA-Z]+)',
        '',
        name
    )

    return name


def extract_product_name(text):
    lines = get_lines(text)
    candidates = []

    #print(lines)
    for line in lines[:10]:  # names usually near top
        if PRODUCT_NAME_REGEX.match(line):
            candidates.append((score_name(line), line))

    if not candidates:
        return None

    candidates.sort(reverse=True)
    clenaed_name = clean_product_name(candidates[0][1])
    cleaned_name = clean_product_prefix(clenaed_name)
    
    return cleaned_name
````

````python
import re

def extract_ml_value(text):
    """
    Extract ML value from text, handling various formats including:
    - 250ml, 25O ml, 250 ml, 2 5 0 ml
    - 24 X 200 ml (returns 200, not 24)
    """
    text = re.sub(r'(\d+)\s*[O]+\s*(\d*)\s*([mM])', lambda m: m.group(1) + '0' * m.group(0).count('O') + m.group(2) + ' ' + m.group(3), text)
    text = re.sub(r'([O]+)\s*(\d+)\s*([mM])', lambda m: '0' * m.group(0).count('O') + m.group(2) + ' ' + m.group(3), text)
    

    
    ML_REGEX = re.compile(
        r'(?<!\d)(\d[\s\d]*\d|\d)\s*[mM]\s*[lL1I|]',
        re.IGNORECASE
    )
    
    matches = ML_REGEX.findall(text)
    
    if not matches:
        return None
    
    # Process matches to remove spaces and convert to int
    ml_values = []
    for match in matches:
        # Remove all spaces from the matched digits
        cleaned = match.replace(' ', '')
        try:
            value = int(cleaned)
            # Filter out unreasonably large values (likely not ML measurements)
            if 10 <= value <= 10000:  # reasonable range for ML values
                ml_values.append(value)
        except ValueError:
            continue
    
    # Return the largest reasonable value found
    # (handles cases like "24 X 200 ml" by returning 200)
    return max(ml_values) if ml_values else None
````

````python
import cv2
import numpy as np
import torch
import albumentations as A
import os

transform = A.Compose([
    A.Resize(256, 256),
])

def preprocess_image(image_bytes):
    """
    Preprocess image from bytes for model input
    
    Args:
        image_bytes: Image data as bytes
    
    Returns:
        torch.Tensor: Preprocessed image tensor
    """
    # Decode bytes to image
    nparr = np.frombuffer(image_bytes, np.uint8)
    image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Apply transformations
    augmented = transform(image=image)
    image = augmented["image"]
    image = image.transpose(2, 0, 1)  # HWC → CHW
    image = torch.tensor(image, dtype=torch.float32) / 255.0
    return image.unsqueeze(0)


def extract_text_regions_from_mask(
    image_bytes,
    pred_mask,
    min_area=500,
    min_width=20,
    min_height=10,
    kernel_size=(5, 5)
):
    """
    Extracts text regions from image bytes using a segmentation mask.
    
    Args:
        image_bytes: Image data as bytes
        pred_mask: Predicted mask from model (numpy array)
        min_area: Minimum contour area to keep
        min_width: Minimum width of bounding box
        min_height: Minimum height of bounding box
        kernel_size: Morphological kernel size for closing operation
    
    Returns:
        crops (list): List of cropped image regions (BGR format)
        boxes (list): Corresponding bounding boxes (x, y, w, h)
    """
    # Decode bytes to image
    nparr = np.frombuffer(image_bytes, np.uint8)
    image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    
    if image is None:
        raise ValueError("Could not decode image from bytes")
    
    # Resize mask to match original image size
    mask_resized = cv2.resize(
        pred_mask.astype(np.uint8) * 255,
        (image.shape[1], image.shape[0])
    )
    
    # Mask cleanup (CRITICAL)
    # Remove salt-and-pepper noise
    mask_resized = cv2.medianBlur(mask_resized, 5)
    
    # Merge broken characters / words
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernel_size)
    mask_resized = cv2.morphologyEx(mask_resized, cv2.MORPH_CLOSE, kernel)
    
    # Find contours (text regions)
    contours, _ = cv2.findContours(
        mask_resized,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )
    
    crops = []
    boxes = []
    
    # Filter and crop
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area < min_area:
            continue
        
        x, y, w, h = cv2.boundingRect(cnt)
        if w < min_width or h < min_height:
            continue
        
        crop = image[y:y + h, x:x + w]
        if crop.size == 0:
            continue
        
        crops.append(crop)
        boxes.append((x, y, w, h))
    
    return crops, boxes


def complete_pipeline(image_bytes, model, DEVICE):
    """
    Complete pipeline with robust output handling
    """
    # Step 1: Preprocess image for model
    image_tensor = preprocess_image(image_bytes).to(DEVICE)
    
    # Step 2: Get model prediction
    model.eval()
    with torch.no_grad():
        output = model(image_tensor)
        
        # Handle different output types
        if isinstance(output, dict):
            # For models that return dict (e.g., FCN, DeepLabV3)
            output = output['out']
        
        # Get the prediction
        pred = output.squeeze().cpu().numpy()
        
        # Handle multi-class outputs
        if len(pred.shape) == 3:  # (num_classes, H, W)
            # Take argmax for multi-class or first channel for binary
            if pred.shape[0] > 1:
                pred_mask = np.argmax(pred, axis=0).astype(np.uint8)
            else:
                pred_mask = pred[0]
        else:  # (H, W)
            pred_mask = pred
        
        # Binarize the mask
        pred_mask = (pred_mask > 0.5).astype(np.uint8)
    
    # Step 3: Extract text regions
    crops, boxes = extract_text_regions_from_mask(
        image_bytes,
        pred_mask,
        min_area=500,
        min_width=70,
        min_height=10,
        kernel_size=(5, 5)
    )
    
    return crops
````

````python
import cv2
import numpy as np
import torch
import albumentations as A
import matplotlib.pyplot as plt

transform = A.Compose([
    A.Resize(256, 256),
])

def preprocess_image(image_bytes):
    """Preprocess image bytes for model"""
    nparr = np.frombuffer(image_bytes, np.uint8)
    image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    augmented = transform(image=image)
    image = augmented["image"]
    image = image.transpose(2, 0, 1)
    image = torch.tensor(image, dtype=torch.float32) / 255.0
    return image.unsqueeze(0)

def extract_text_regions_from_mask(image_bytes, pred_mask, min_area=500, min_width=50, min_height=50, kernel_size=(5, 5)):
    """Extract crops from image using mask"""
    nparr = np.frombuffer(image_bytes, np.uint8)
    image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    
    if image is None:
        raise ValueError("Could not decode image")
    
    mask_resized = cv2.resize(pred_mask.astype(np.uint8) * 255, (image.shape[1], image.shape[0]))
    mask_resized = cv2.medianBlur(mask_resized, 5)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernel_size)
    mask_resized = cv2.morphologyEx(mask_resized, cv2.MORPH_CLOSE, kernel)
    
    contours, _ = cv2.findContours(mask_resized, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    crops = []
    boxes = []
    
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area < min_area:
            continue
        
        x, y, w, h = cv2.boundingRect(cnt)
        if w < min_width or h < min_height:
            continue
        
        crop = image[y:y + h, x:x + w]
        if crop.size == 0:
            continue
        
        crops.append(crop)
        boxes.append((x, y, w, h))
    
    return crops, boxes

def complete_pipeline(image_bytes, model, DEVICE):
    """Complete pipeline"""
    image_tensor = preprocess_image(image_bytes).to(DEVICE)
    
    model.eval()
    with torch.no_grad():
        output = model(image_tensor)
        
        if isinstance(output, dict):
            output = output['out']
        
        pred_mask = output.squeeze().cpu().numpy()
        
        if len(pred_mask.shape) == 3:
            if pred_mask.shape[0] > 1:
                pred_mask = np.argmax(pred_mask, axis=0)
            else:
                pred_mask = pred_mask[0]
        
        pred_mask = (pred_mask > 0.5).astype(np.uint8)
    
    crops, boxes = extract_text_regions_from_mask(image_bytes, pred_mask)
    return crops

# Simple visualization
def show_crops(crops):
    for i, crop in enumerate(crops):
        plt.figure(figsize=(10, 3))
        plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
        plt.title(f"Crop {i+1}")
        plt.axis('off')
        plt.show()

# Usage:
with open("img.png", "rb") as f:
    image_bytes = f.read()

````
